```python
#Importar librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#- Importa ambos conjuntos de datos usando Pandas.
df_prices = pd.read_csv('stocks_daily_prices.csv')
df_returns = pd.read_csv('stocks_daily_returns.csv')
```

Con Matplotlib, representa diagramas de líneas que muestren los precios diarios de las 4 acciones en una sola cifra

```python
df_prices.plot(x='Date', y=['AAPL','JPM', 'PG', 'UAL'], linewidth= 3, figsize=(10, 6), ylabel='Precio en USD', xlabel='Fecha', title="Precios diarios de acciones")
```

Con Matplotlib, representa 4 precios diarios de acciones en varias suplots.

```python
df_prices.plot(x='Date', y=['AAPL','JPM', 'PG', 'UAL'], grid= True, linewidth= 3, figsize=(10, 10), subplots=True, title= 'Precios diarios de acciones', ylabel="Precio en USD")
plt.xlabel('Fecha')
plt.show()
```

Usando Matplotlib, representa los 4 plots en subplots una al lado de la otra (todas las figuras en una fila).

```python
plt.figure(figsize=(20, 4))

plt.subplot(1, 4, 1)
plt.plot(df_prices['AAPL'], linewidth= 2, label='AAPL', color= 'b')
plt.grid()
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.title('Precios diarios de AAPL')
plt.legend()

plt.subplot(1, 4, 2)
plt.plot(df_prices['JPM'], linewidth= 2, label='JPM', color= 'y')
plt.grid()
plt.xlabel('Fecha')
plt.title('Precios diarios de JPM')
plt.legend()

plt.subplot(1, 4, 3)
plt.plot(df_prices['PG'], linewidth= 2, label='PG', color= 'g')
plt.grid()
plt.xlabel('Fecha')
plt.title('Precios diarios de PG')
plt.legend()

plt.subplot(1, 4, 4)
plt.plot(df_prices['UAL'], linewidth= 2, label='UAL', color= 'r')
plt.grid()
plt.xlabel('Fecha')
plt.title('Precios diarios de UAL')
plt.legend()

plt.show()
```

Con Matplotlib, representa el diagrama de dispersión entre los rendimientos diarios de Apple y JP Morgan.

```python
plt.figure(figsize = (15, 10))
df_returns.plot.scatter('AAPL','JPM', alpha= 0.7)
plt.grid()
plt.xlabel('Rendimiento diario de AAPL')
plt.ylabel('Rendimiento diario de JPM')
plt.title('Diagrama de dispersión entre los rendimientos diarios de Apple y JP Morgan')
```

Utilizando Seaborn, representa un diagrama de dispersión similar entre los rendimientos diarios de Apple y JP Morgan.

```python
sns.scatterplot(x='AAPL', y='JPM', data=df_returns)
plt.grid()
plt.xlabel('Rendimiento diario de AAPL')
plt.ylabel('Rendimiento diario de JPM')
plt.title('Diagrama de dispersión entre los rendimientos diarios de Apple y JP Morgan')
```

Supon que decides volverse alcista sobre AAPL y asignas el 70% de sus activos. También decides dividir en partes iguales el resto de sus activos en otras acciones (JPM, PG y UAL). Con Matplotlib, representa un gráfico circular que muestre estas

```python
#Generamos un DataFrame con el porcentage de inversiones para cada accion
inversiones_df = pd.DataFrame(data={'allocation %':[70, 10, 10, 10]}, index=["AAPL", "JPM", "PG", "UAL"])

inversiones_df
```

```python
#Generamos el grafico de pastel con el portfolio de inversiones
inversiones_df.plot.pie(y="allocation %", figsize=(7, 7), explode= (0.1,0,0,0))
plt.title('Portfolio de inversiones')
plt.ylabel('')
plt.show()
```

Usando Matplotlib, representa el histograma para los retornos de United Airlines y P&G usando 40 divisiones con color rojo. Muestra la media y la desviación estándar para ambas acciones en la parte superior de la figura. ¿Qué concluyes de este gráfico?

```python
#Obtenemos la media y la desviación estandar para ambas acciones
media_ual = df_returns['UAL'].mean()
std_ual = df_returns['UAL'].std()

media_pg = df_returns['PG'].mean()
std_pg = df_returns['PG'].std()

num_bins = 40
plt.figure(figsize=(10, 6))

#Histograma para United Airlines
plt.subplot(2, 1, 1)
plt.hist(df_returns['UAL'], bins=num_bins, color='red', alpha=0.7)
plt.axvline(media_ual, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(media_ual + std_ual, color='g', linestyle='dashed', linewidth=2, label='Desviación Estándar')
plt.axvline(media_ual - std_ual, color='g', linestyle='dashed', linewidth=2)
plt.title('Histograma de retornos de United Airlines, media: ' + str(round(media_ual, 2)) + ' std: ' + str(round(std_ual, 2)))
plt.xlabel('Rendimiento diario')
plt.ylabel('Frecuencia')
plt.legend()

#Histogrma para PG
plt.subplot(2, 1, 2)
plt.hist(df_returns['PG'], bins=num_bins, color='red', alpha=0.7)
plt.axvline(media_pg, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(media_pg + std_pg, color='g', linestyle='dashed', linewidth=2, label='Desviación Estándar')
plt.axvline(media_pg - std_pg, color='g', linestyle='dashed', linewidth=2)
plt.title('Histograma de retornos de P&G, media: '+ str(round(media_pg, 2)) + ' std: ' + str(round(std_pg, 2)))
plt.xlabel('Rendimiento diario')
plt.ylabel('Frecuencia')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (20, 15));

mu_UAL = round(df_returns['UAL'].mean(), 2)
sigma_UAL = round(df_returns['UAL'].std(), 2)


mu_PG = round(df_returns['PG'].mean(), 2) # mean of distribution
sigma_PG = round(df_returns['PG'].std(), 2) # standard deviation of distribution


df_returns[['PG','UAL']].plot.hist(bins = 60, alpha = 0.7, figsize = (15, 8));
plt.grid();
plt.ylabel('Probabilidad')


plt.title('Histograma: P&G mu=' + str(mu_PG) + ', P&G sigma=' +str(sigma_PG) + ', United Airlines mu=' +str(mu_UAL) + ', United Airlines sigma=' +str(sigma_UAL) );
```

```python
plt.figure(figsize = (12, 10))
#Excluir la columna 'Date' al calcular la matriz de correlación
cm = df_returns.drop(columns=['Date']).corr()
sns.heatmap(cm, annot = True)
```

Representa un gráfico 3D que muestre todos los retornos diarios de JPM, AAPL y UAL [Se requiere investigación externa].

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(df_returns['JPM'], df_returns['AAPL'], df_returns['UAL'], c='r', marker='o', s=100)
ax.set_xlabel('JPM')
ax.set_ylabel('AAPL')
ax.set_zlabel('UAL')
plt.show()
```
