# Funciones con python

```python
hr_df=pd.read_csv("Human_Resources.csv")
hr_df
```

```python
#Supongamos que el daily rate ha aumentado un 10%.
#Definamos una funciÃ³n que aumenta el patrimonio de todos los clientes en un valor fijo del 10% (por simplicidad)
def daily_rate_aum(daily_rate):
    return daily_rate*1.1 #supongamos que el patrimonio neto aumenta un 10%
```

```python
#podemos aplicar una funcion al DataFrame
hr_df["DailyRate"]=hr_df["DailyRate"].apply(daily_rate_aum)
hr_df["DailyRate"]
```

#Practica 7:
###*Definir una funciÃ³n que duplicque la tasa diaria y aÃ±ada $100.
###*Aplicar la funcion al DataFrame.
###*Calcular la tasa diaria total actualizada de todos los empleados combinados.

```python
hr_df=pd.read_csv("Human_Resources.csv")
hr_df
```

```python
#Definir una funciÃ³n que duplicque la tasa diaria y aÃ±ada $100.
def daily_rate_aum(daily_rate):
    return daily_rate*2+100
```

```python
#Aplicar la funcion al DataFrame.
hr_df["DailyRate"]=hr_df["DailyRate"].apply(daily_rate_aum)
hr_df["DailyRate"]
```

```python
#Calcular la tasa diaria total actualizada de todos los empleados combinados.
hr_df["DailyRate"].sum()
```
